```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
```

```python
#總資料組數
N = 1000000
```

```python
ve_dune = []
vebar_dune = []
vu_dune = []
vubar_dune = []
ve_t2hk = []
vu_t2hk = []
vebar_t2hk = []
vubar_t2hk = []
theta12 = []
theta13 = []
theta23 = []
delta = []
sdm = []
ldm = []
octant = []
cpv = []
mo = []
```

```python
#Import txt檔
f = open("n1000000_0804_all_flat.txt")
```

```python
z = [0,1,2,3]
z[0:0]
```

```python
for i in range(N):
    s = f.readline().split()
    array = []
    for j in range(len(s)) :
        array.append(float(s[j]))
    ve_dune.append(array[0:66])
    vebar_dune.append(array[66:132])
    vu_dune.append(array[132:198])
    vubar_dune.append(array[198:264])
    ve_t2hk.append(array[264:272])
    vu_t2hk.append(array[272:284])
    vebar_t2hk.append(array[284:292])
    vubar_t2hk.append(array[292:304])
    theta12.append(array[304])
    theta13.append(array[305])
    theta23.append(array[306])
    delta.append(array[307])
    sdm.append(array[308])
    ldm.append(array[309])
    octant.append(array[310])
    cpv.append(array[311])
    mo.append(array[312])
    

```

```python
#輸出全部資料為npz檔
np.savez("n1000000_0910_all_flat",
         ve_dune = ve_dune,
         vebar_dune = vebar_dune,
         vu_dune = vu_dune,
         vubar_dune = vubar_dune,
         ve_t2hk = ve_t2hk,
         vu_t2hk = vu_t2hk,
         vebar_t2hk = vebar_t2hk,
         vubar_t2hk= vubar_t2hk,
         theta12 = theta12,
         theta13 = theta13,   
         theta23 = theta23,   
         delta = delta,
         sdm = sdm,   
         ldm = ldm,  
         octant = octant,
         cpv = cpv,   
         mo = mo  
        )
```
