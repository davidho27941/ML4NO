```python
%%time
def preprocessing(data,column=2):
    tmp = data.reshape(len(data),column,36)
    tmp = np.array([ element.T for element in tmp])
    
    return tmp

def Contour(resolution_dictinary, model_dictinary, parameters="delta", model="model_1"):
    logging.info("==========================================================")
    logging.info("===================== Model: {} =====================".format(model))
    logging.info("==========================================================")
    logging.info("\n")
    
    if parameters == "delta":
        first_axis, second_axis = 0, 1
    elif parameters == "theta23":
        first_axis, second_axis = 2, 3
        
        
    for i, std_scale in enumerate(tqdm(resolution_dictinary)):
        

        
        if model == "model_1" or model == "model_2":
#             prediction_asimov = model_dictinary[std_scale].predict(data_asimov)[0][0]
            tmp = model_dictinary[std_scale].predict(mc_data)[:,0]
            
        elif model == "model_3" or model == "model_4" or model == "model_5":
#             prediction_asimov = model_dictinary[std_scale].predict([data_asimov,data_asimov[:,:72]-data_asimov[:,72:]])[0][0]
            tmp = model_dictinary[std_scale].predict([mc_data,mc_data[:,:72]-mc_data[:,72:]])#[:,0]            
    
            """
            TEST
            """
            tmp = np.arctan2(tmp[:,0],tmp[:,1]) # 10/26 modified 
            tmp = np.where(tmp >= 0 , tmp, 2*np.pi+tmp) # 10/26 modified 
            tmp = tmp/np.pi*180 # 10/26 modified 
            
            
        elif model == "model_new_1" or model == "model_new_2": # 11/05 modified 
                
            mc_data_tmp = mc_data.reshape(len(mc_data),4,36)
            mc_data_tmp = np.array([ element.T for element in mc_data_tmp])
            tmp = model_dictinary[std_scale].predict(mc_data_tmp)
            tmp = np.arctan2(tmp[:,first_axis],tmp[:,second_axis])
            tmp = np.where(tmp >= 0 , tmp, 2*np.pi+tmp) 
            tmp = tmp/np.pi*180 
            
            
        elif model == "model_new_3" or model == "model_new_4": # 11/08 modified 
            
            if parameters == "delta":
                
                mc_data_tmp = mc_data.reshape(len(mc_data),4,36)
                mc_data_tmp = np.array([ element.T for element in mc_data_tmp])
                tmp = model_dictinary[std_scale].predict(mc_data_tmp)[0]
                tmp = np.arctan2(tmp[:,0],tmp[:,1])
                tmp = np.where(tmp >= 0 , tmp, 2*np.pi+tmp) 
                tmp = tmp/np.pi*180 
                
                
            elif parameters == "theta23":
                
                
                mc_data_tmp = mc_data.reshape(len(mc_data),4,36)
                mc_data_tmp = np.array([ element.T for element in mc_data_tmp])
                tmp = model_dictinary[std_scale].predict(mc_data_tmp)[1]
                tmp = np.where(tmp > 0 , tmp, 2*np.pi+tmp) 
                tmp = tmp/np.pi*180
                
        elif model == "model_new_5_s" : # 11/08 modified 
            
            mc_data_tmp = scaler.transform(mc_data)
            appearance_poisson = preprocessing(mc_data_tmp[:,:72],column=2)
            disappearance_poisson = preprocessing(mc_data_tmp[:,72:],column=2)
            
            
            if parameters == "delta":
                tmp = model_dictinary[std_scale].predict([appearance_poisson,disappearance_poisson])[0]
                tmp = np.arctan2(tmp[:,0],tmp[:,1])
                tmp = np.where(tmp >= 0 , tmp, 2*np.pi+tmp) 
                tmp = tmp/np.pi*180 
                
                
            elif parameters == "theta23":
            
                tmp = model_dictinary[std_scale].predict([appearance_poisson,disappearance_poisson])[1]
                tmp = np.where(tmp >= 0 , tmp, 2*np.pi+tmp) 
                tmp = tmp/np.pi*180 

        elif model == "model_new_6":
            
            appearance_poisson = preprocessing(mc_data[:,:72],column=2)
            disappearance_poisson = preprocessing(mc_data[:,72:],column=2)
            poisson_spetcrum_diff = np.column_stack([(mc_data[:,:36] - mc_data[:,36:72]),(mc_data[:,72:108] - mc_data[:,108:144])]) 
            
            if parameters == "delta":
                tmp = model_dictinary[std_scale].predict([appearance_poisson,disappearance_poisson,mc_data,poisson_spetcrum_diff])[0]
                tmp = np.arctan2(tmp[:,0],tmp[:,1])
                tmp = np.where(tmp >= 0 , tmp, 2*np.pi+tmp) 
                tmp = tmp/np.pi*180 
                
                
            elif parameters == "theta23":
            
                tmp = model_dictinary[std_scale].predict([appearance_poisson,disappearance_poisson,mc_data,poisson_spetcrum_diff])[1]
                tmp = np.where(tmp >= 0 , tmp, 2*np.pi+tmp) 
                tmp = tmp/np.pi*180 

            
        else:
            logging.info("Please Check Input Model!!!")
            raise ValueError("Please Check Input Model!!!")

        resolution_dictinary[std_scale]["poission_prediction"] = tmp
    
    return resolution_dictinary
    
    

theta23_prediction = {}
delta_prediction = {}

for i in scale_range:
    
    if i == -1:  
        theta23_prediction.update({"asimov": {"poission_prediction":0},}) 
        delta_prediction.update({"asimov": {"poission_prediction":0},}) 
        
    else:
        theta23_prediction.update({"std_scale_"+str(i): {"poission_prediction":0},}) 
        delta_prediction.update({"std_scale_"+str(i): {"poission_prediction":0},}) 


    
theta23_prediction = Contour(theta23_prediction, model_theta23, parameters="theta23", model=nth_model)
        
delta_prediction = Contour(delta_prediction, model_delta, parameters="delta", model=nth_model)
    
    
```
