```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
```

```python
#filename = "best_fit_spectrum_deltacp_theta23_DUNE"
#filename = "best_fit_spectrum_deltacp_theta23_T2HK"
#filename = "best_fit_spectrum_deltacp_theta23_DUT2"
#filename = "best_fit_spectrum_deltacp_theta23_DUNE_NO"
filename = "best_fit_spectrum_deltacp_theta23_DUNE_IO"
```

```python
#Import txt檔
f = open("/docker_workplace/globes-3.2.18/examples/" + filename+".dat")
```

```python
total_num = 100
```

```python
theta12_true = []
theta13_true = []
theta23_true = []
delta_true = []
sdm_true = []
ldm_true = []

chi2 = []

theta12_fit = []
theta13_fit = []
theta23_fit = []
delta_fit = []
sdm_fit = []
ldm_fit = []

event_run_time = []

ve_dune_poisson = []
vebar_dune_poisson = []
vu_dune_poisson = []
vubar_dune_poisson = []
ve_t2hk_poisson= []
vu_t2hk_poisson = []
vebar_t2hk_poisson = []
vubar_t2hk_poisson = []

for i in range(total_num):
    s = f.readline().split()
    array = []
    for j in range(len(s)) :
        array.append(float(s[j]))
    if(len(array)>0):
        theta12_true.append(array[0])
        theta13_true.append(array[1])
        theta23_true.append(array[2])
        delta_true.append(array[3])
        sdm_true.append(array[4])
        ldm_true.append(array[5])

        chi2.append(array[6])

        theta12_fit.append(array[7])
        theta13_fit.append(array[8])
        theta23_fit.append(array[9])
        delta_fit.append(array[10])
        sdm_fit.append(array[11])
        ldm_fit.append(array[12])
        
        event_run_time.append(array[13])
        
        ve_dune_poisson.append(array[0+13:66+13])
        vebar_dune_poisson.append(array[66+13:132+13])
        vu_dune_poisson.append(array[132+13:198+13])
        vubar_dune_poisson.append(array[198+13:264+13])
        ve_t2hk_poisson.append(array[264+13:272+13])
        vu_t2hk_poisson.append(array[272+13:284+13])
        vebar_t2hk_poisson.append(array[284+13:292+13])
        vubar_t2hk_poisson.append(array[292+13:304+13])
    else:
        break    
```

```python
#輸出全部資料為npz檔
np.savez("/docker_workplace/globes-3.2.18/examples/python/output/"+filename,
        theta12_true = theta12_true,
        theta13_true = theta13_true,
        theta23_true = theta23_true,
        delta_true = delta_true,
        sdm_true = sdm_true,
        ldm_true = ldm_true,

        chi2 = chi2,

        theta12_fit = theta12_fit,
        theta13_fit = theta13_fit,
        theta23_fit = theta23_fit,
        delta_fit = delta_fit,
        sdm_fit = sdm_fit,
        ldm_fit = ldm_fit,
        
        event_run_time = event_run_time, 
         
        ve_dune_poisson = ve_dune_poisson,
        vebar_dune_poisson = vebar_dune_poisson,
        vu_dune_poisson = vu_dune_poisson,
        vubar_dune_poisson = vubar_dune_poisson,
        ve_t2hk_poisson= ve_t2hk_poisson,
        vu_t2hk_poisson = vu_t2hk_poisson,
        vebar_t2hk_poisson = vebar_t2hk_poisson,
        vubar_t2hk_poisson = vubar_t2hk_poisson,
        )
```
